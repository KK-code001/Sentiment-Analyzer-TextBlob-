```python
import json
from textblob import TextBlob
```

```python
def textcleaner(message):
  message=message.lower()
  punctuation=".,!?;:'\"()-[]{}"
  for p in punctuation:
      message=message.replace(p,"")
  return message.strip()

def sentiment_classify(score):
    if score>=0.80:
        return "Very Positive"
    elif score>=0.50:
        return "Positive"
    elif score>=0.20:
        return "Neutral-Positive"
    elif score>-0.20:
        return "Neutral"
    elif score>-0.50:
        return "Neutral-Negative"
    elif score>-0.80:
        return "Negative"
    else:
        return "Very Negative"

def sentiment_analyze(message):
  cleaned=textcleaner(message)
  blob=TextBlob(cleaned)
  score=blob.sentiment.polarity
  sub=blob.sentiment.subjectivity
  print(f"Sentiment Score={score} Subjectivity Score={sub}")
  sentiment=sentiment_classify(score)
  return {
        "Original Text":message,
        "Cleaned Text":cleaned,
        "Polarity":score,
        "Subjectivity":sub,
        "Sentiment":sentiment
  }

def writetojson(res):
  try:
    with open("analysis_data.json","x") as file:
        json.dump([res],file,indent=4)
  except FileExistsError:
      with open("analysis_data.json","r") as file:
        try:
          data=json.load(file)
        except json.JSONDecodeError:
          data=[]
        if not isinstance(data,list):
            data=[data]
        data.append(res)
      with open("analysis_data.json","w") as file:
        json.dump(data,file,indent=4)
```

```python
def main():
  filename=input("Enter filename: ")
  try:
    with open(filename,"r") as file:
      data=file.read()
  except FileNotFoundError:
    print("File not found")
    return
  res=sentiment_analyze(data)
  print(res)
  writetojson(res)

main()
```
